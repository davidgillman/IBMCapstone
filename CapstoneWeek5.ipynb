Merge Dumpling location dataframe (df1) with centers locations to compute Euclidean Distance

```python
dfDumpling = df1
dfDumpling['Latitude1'] = 43.660971
dfDumpling['Longitude1'] = -79.390803
dfDumpling['Latitude2'] = 43.778451
dfDumpling['Longitude2'] = -79.227678
dfDumpling['Latitude3'] = 43.681909
dfDumpling['Longitude3'] = -79.527637
dfDumpling['Latitude4'] = 43.742878
dfDumpling['Longitude4'] = -79.420558
dfDumpling['Latitude5'] = 43.721159
dfDumpling['Longitude5'] = -79.317069
dfDumpling
```

```python
dfDumpling.describe()
```

```python
# Calculate Distances from each center
dfDumpling['Dist1'] = (((dfDumpling['Latitude']-dfDumpling['Latitude1'])**2)+((dfDumpling['Longitude']-dfDumpling['Longitude1'])**2))**0.5
dfDumpling['Dist2'] = (((dfDumpling['Latitude']-dfDumpling['Latitude2'])**2)+((dfDumpling['Longitude']-dfDumpling['Longitude2'])**2))**0.5
dfDumpling['Dist3'] = (((dfDumpling['Latitude']-dfDumpling['Latitude3'])**2)+((dfDumpling['Longitude']-dfDumpling['Longitude3'])**2))**0.5
dfDumpling['Dist4'] = (((dfDumpling['Latitude']-dfDumpling['Latitude4'])**2)+((dfDumpling['Longitude']-dfDumpling['Longitude4'])**2))**0.5
dfDumpling['Dist5'] = (((dfDumpling['Latitude']-dfDumpling['Latitude5'])**2)+((dfDumpling['Longitude']-dfDumpling['Longitude5'])**2))**0.5
dfDumpling
```

```python
DistanceOut1 = df1
DistanceOut1['DumplingsClosest1'] = dfDumpling['Dist1'].min()
DistanceOut1['DumplingsClosest2'] = dfDumpling['Dist2'].min()
DistanceOut1['DumplingsClosest3'] = dfDumpling['Dist3'].min()
DistanceOut1['DumplingsClosest4'] = dfDumpling['Dist4'].min()
DistanceOut1['DumplingsClosest5'] = dfDumpling['Dist5'].min()
DistanceOut1['DumplingsAverage1'] = dfDumpling['Dist1'].mean()
DistanceOut1['DumplingsAverage2'] = dfDumpling['Dist2'].mean()
DistanceOut1['DumplingsAverage3'] = dfDumpling['Dist3'].mean()
DistanceOut1['DumplingsAverage4'] = dfDumpling['Dist4'].mean()
DistanceOut1['DumplingsAverage5'] = dfDumpling['Dist5'].mean()
DistanceOut1.head()
```
