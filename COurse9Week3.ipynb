```python
# import packages needed for clustering and mapping

from sklearn.cluster import KMeans
import folium
```

```python
#plot Toronto on Map

latitude = 43.70
longitude = -79.38
GTA_map = folium.Map(location=[latitude, longitude], zoom_start=11)
GTA_map
```

```python
# add neighborhood markers
hoods = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df3.Latitude, df3.Longitude):
    hoods.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
GTA_map.add_child(hoods)
```

I don't see any obvious clustering.  Just some increasing denisty towards downtown.  So I will try to find some clusters.  I will use K-means and set the cluster to 5 just to see what the system comes up with.

First seperate Latitude and Longitude into a seperate, skinnier data frame.

```python
df4 = df3[['Latitude', 'Longitude']]
df4.columns = ['X', 'Y']
df4
```

```python
#  convert X and Y (Lat and Long) to array for Kmeans algorithm ingestion

df4_array = df4.values

# Define cluster and fit data

k_means = KMeans(init = "k-means++", n_clusters = 5, n_init = 12)
k_means.fit(df4_array)
```

```python
#  Get centroids to plot on new map 

k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```

```python
# Create New Map

latitude = 43.70
longitude = -79.38
GTA_C_map = folium.Map(location=[latitude, longitude], zoom_start=11)
GTA_C_map
```

```python
# convert array of centers to data frame with Latitude and Longitude

df5 = pd.DataFrame(k_means_cluster_centers)
df5.columns = ['Latitude', 'Longitude']
df5
```

```python
# add neighborhood markers
hoodcenters = folium.map.FeatureGroup()


# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df5.Latitude, df5.Longitude):
    hoodcenters.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='red',
            fill=True,
            fill_color='white',
            fill_opacity=0.6
        )
    )

# add incidents to map
GTA_C_map.add_child(hoodcenters)
```
